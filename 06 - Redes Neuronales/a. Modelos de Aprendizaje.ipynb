## MODELOS DE APRENDIZAJE

### Aprendizaje Máquina (Machine Learning)

El **Aprendizaje Máquina (Machine Learning, ML)** es un campo de estudio que permite que una computadora tenga una capacidad de aprendizaje o experiencia, para con ello poder tomar decisiones y hacer pronósticos.

Permite a los equipos de cómputo realizar la toma de decisiones de la manera más óptima y basadas totalmente en los datos y su aprendizaje, sin la necesidad de ser programadas de manera explícita para llevar a cabo cierta acción. Estos programas o algoritmos se diseñan de manera que exista un proceso de aprendizaje que pueda ser mejorado con el tiempo y con la adquisición de nuevos datos que contribuyen a robustecer el proceso.

#### El Aprendizaje Máquina puede ser dividido en 3 categorías: 

#### 1. Aprendizaje Supervisado (Supervised Learning)
En el aprendizaje supervisado, se cuenta con datos ya etiquetados que contienen un dato de entrada ***X*** y su etiqueta ***Y***. El objetivo es realizar un mapeo entre la variable de entrada (***X***) a la cual se le conoce como *variable independiente*, con la variable de salida (***Y***) a la cual se le conoce como *variable dependiente*.

Este tipo de aprendizaje puede ser dividido en dos tareas:

* 1. **Regresión (Regression)**: Los problemas de regresión ocurren cuando la variable de salida es continua y de un valor real, por ejemplo, precios, pesos, entre otros.
* 2. **Clasificación (Classification)**: La clasificación se refiere a un problema en el que la variable de salida es una categoría, como *rojo* y *azul*, o *pagado* y *no pagado*.

#### 2. Aprendizaje No Supervisado (Unsupervised Learning)
En el aprendizaje no supervisado, solamente se cuenta con los datos de entrada (***X***) y no se tienen valores correspondientes de salida. El objetivo es modelar la distribución de los datos para poder tener mayor información al respecto y poder aprender de los mismos.

La siguiente figura muestra un ejemplo aplicado a la clasificación de zonas en una imagen satelital:

<br>
<br>
El aprendizaje no supervisado puede ser divido en dos tareas:

* 1. **Agrupamiento (Clustering)**: Un problema de agrupamiento es cuando se requiere identificar los grupos inherentes en los datos, por ejemplo, realizar la agrupación de compradores en base a la manera de realizar las compras.
* 2. **Asociación (Association)**: Un problema que emplea una regla de asociación es aquél donde se requiere identificar las reglas que describen grandes porciones de los datos, por ejemplo, cuando la gente que compra el producto *a* que tiene a comprar también el producto *b*. 

#### 3. Aprendizaje Reforzado (Reinforcement Learning)
En el contexto de la **Inteligencia Artificial**, el aprendizaje reforzado es un tipo de programación dinámica que entrenan un algoritmo a través de un sistema de recompensa y penalización. 

El algoritmo que realiza su proceso de aprendizaje por un método reforzado, al cual se le conoce como *agente*, aprende por medio de la interacción con su ambiente. El *agente* recibe una recompensa cuando se realiza de manera adecuada, y una penalización cuando no se realiza de manera adecuada. El *agente* aprende sin intervención humana al realizar un proceso de maximización de su recompensa y, por ende, de minimización de su penalización.

### Aprendizaje Profundo (Deep Learning)

El **Aprendizaje Profundo (Deep Learning, DL)** es un sub-campo de **Aprendizaje Máquina (Machine Learning)** que se enfoca a los algoritmos inspirados en la estructura y funciones del derebro humano, los cuales se conocen como ***Redes Neuronales Artificiales (Artificial Neural Networks, ANN)***.

#### ¿Por qué existe tanto interés científico en el Aprendizaje Profundo (Deep Learning)?
Las ***Redes Neuronales Artificiales (Artificial Neural Networks, ANN)*** son los bloques básicos de construcción del **Aprendizaje Profundo (Deep Learning)** pero no son modelos nuevos, e incluso son más antiguas que muchos algoritmos de **Aprendizaje Máquina (Machine Learning)**. En su momento de desarrollo inicial, no fueron tan conocidas y demandadas, principalmente por la falta de poder de cómputo que existía, así como la falta de datos para poder tener resultados robustos.  

En el 2012 un equipo de investigadores ganaron un reto de visión computacional organizado por [ImageNet](http://www.image-net.org) empleando ***Redes Neuronales Convolucionales (Convolutional Neural Networks, CNN)*** por vez primera en esta competición, lo cual provocó un incremento en el interés en técnicas de **Aprendizaje Máquina (Machine Learning)** a nivel mundial. Ellos crearon la [Arquitectura AlexNet](https://towardsdatascience.com/alexnet-the-architecture-that-challenged-cnns-e406d5297951) para ***CNN***, la cual alcanzó un error de 15.3%, más del 10% de puntos porcentuales por debajo del segundo lugar. De manera similar, la idea de una ***CNN*** no era novedosa y fue propuesta en el año 1998 por *Yan Le Cun*, quien es considerado un pionero de la inteligencia artificial moderna. En 2012, y en gran parte debido al poder computacional que estaba disponible, así como la cantidad de datos a los que se tenía acceso, las ***Redes Neuronales*** mostraron su verdadero potencial.

#### Aplicaciones del Aprendizaje Profundo (Deep Learning)
Las aplicaciones del **Aprendizaje Profundo (Deep Learning)** son muy vastas, y muchas de las tecnologías de punta actuales emplean estas metodologías para mejorar y robustecer sus desarrollos. Algunos ejemplos son:
* 1. Autos de conducción asistida.
* 2. Búsqueda por voz y asistentes virtuales.
* 3. Traducción de lenguaje natural.
* 4. Identificación de elementos en imágenes.
* 5. Colorización de imágenes en blanco y negro.
* 6. Inteligencia Artificial en videojuegos (*Open Ai Dota Bot, Google Brain Alpha Go*).
* 7. Reconocimiento de objetos en tiempo real en imágenes (Google Lens).
* 8. Identificación de galletas y Chihuahuas.
